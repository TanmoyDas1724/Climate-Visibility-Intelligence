```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/weatherHistory.csv")
```

```python
df_feature = df.copy()
```

```python
df_feature.head()
```

```python
df_feature["Formatted Date"] = pd.to_datetime(
    df_feature["Formatted Date"],
    utc=True
)
```

```python
df_feature["Formatted Date"].dtype
```

```python
df_feature["Year"] = df_feature["Formatted Date"].dt.year
df_feature["Month"] = df_feature["Formatted Date"].dt.month
df_feature["Day"] = df_feature["Formatted Date"].dt.day
df_feature["Hour"] = df_feature["Formatted Date"].dt.hour
df_feature["DayOfWeek"] = df_feature["Formatted Date"].dt.dayofweek
```

```python
df_feature[["Formatted Date","Year","Month","Day","Hour","DayOfWeek"]].head()
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Autumn"
```

```python
df_feature["Season"] = df_feature["Month"].apply(get_season)
```

```python
df_feature[["Month", "Season"]].head(10)
```

```python
df_feature["Season"].value_counts()
```

```python
df_feature["IsWeekend"] = df_feature["DayOfWeek"].apply(
    lambda x: 1 if x >= 5 else 0
)
```

```python
df_feature[
    ["DayOfWeek", "IsWeekend"]
].head(15)
```

```python
df_feature["IsWeekend"].value_counts()
```

```python
df_feature.select_dtypes(include="str").columns
```

```python
for col in df_feature.select_dtypes(include="str").columns:
    print(f"{col}: {df_feature[col].nunique()} unique values")
```

```python
df_feature = df_feature.drop(columns=["Daily Summary"])
```

```python
df_feature.select_dtypes(include="str").columns
```

```python
for col in df_feature.select_dtypes(include="str").columns:
    print(f"{col}: {df_feature[col].nunique()} unique values")
```

```python
df_feature = pd.get_dummies(
    df_feature,
    columns=["Summary", "Precip Type", "Season"],
    dtype=int
)
```

```python
df_feature.head()
```

```python
df_feature.shape
```

```python
df_feature.drop(columns=["Formatted Date"], inplace=True)
```

```python
df_feature.to_csv(
    "../data/processed/weather_feature_engineered.csv",
    index=False
)
```
