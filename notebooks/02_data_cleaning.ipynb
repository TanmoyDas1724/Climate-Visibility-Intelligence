# Climate Visibility Intelligence Dashboard

## Notebook 02 - Data Cleaning

### Objective

Clean the raw weather dataset before feature engineering and model building.

Tasks:

- Convert date column
- Handle missing values
- Remove unnecessary features
- Fix invalid values
- Save cleaned dataset

````python
import pandas as pd

df = pd.read_csv("../data/raw/weatherHistory.csv")
````

````python
df_clean = df.copy()
````

````python
df_clean.head()
````

````python
df_clean["Formatted Date"] = pd.to_datetime(
    df_clean["Formatted Date"],
    utc=True
)
````

````python
df_clean["Formatted Date"].dtype
````

````python
print("Start Date :", df_clean["Formatted Date"].min())
print("End Date   :", df_clean["Formatted Date"].max())
````

````python
df["Loud Cover"].value_counts()
````

````python
df_clean["Loud Cover"].nunique()
````

````python
df_clean.drop(columns=["Loud Cover"], inplace=True)
````

````python
df_clean.columns
````

````python
df_clean.shape
````

````python
df_clean.isnull().sum()
````

````python
df_clean[df_clean["Precip Type"].isnull()].head()
````

````python
missing_percentage = (
    df_clean["Precip Type"].isnull().sum()
    / len(df_clean)
) * 100

print(f"Missing Percentage: {missing_percentage:.2f}%")
````

````python
df_clean["Precip Type"].value_counts(dropna=False)
````

````python
df_clean.dropna(subset=["Precip Type"], inplace=True)
````

````python
df_clean.isnull().sum()
````

````python
df_clean.shape
````

````python
df_clean.duplicated().sum()
````

````python
df_clean.drop_duplicates(inplace=True)
````

````python
df_clean.duplicated().sum()
````

````python
df_clean.shape
````

````python
df_clean.to_csv(
    "../data/processed/weatherHistory_cleaned.csv",
    index=False
)
````

# Notebook 02 Completed

### Tasks Completed

- Data type conversion
- DateTime standardization (UTC)
- Removed zero variance feature
- Handled missing values
- Removed duplicate rows
- Saved cleaned dataset
